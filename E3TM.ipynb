a) Primero escribimos un algoritmo para calcular el exponente de Liapunov de funciones con un parámetro.

```python
# Definimos la familia de funciones que dependen de un parámetro r.
def f(x,r):
    return r * np.sin(np.pi * x)
```

```python
# Esta función calcula la derivada de las funciones en la familia anterior, en un punto x.
def Df(x,r):
    h = 1e-5
    return (f(x+h,r)-f(x,r))/h
```

```python
# Aquí definimos la función que nos da el exponente de Lyapunov en función de una condición inicial x0,
# una función de la familia anterior determinada por el parámetro r y una n que determina el tamaño de 
# la suma en la fórmula para el exponente.

def lyapunov_exp(x0,n,r):
    x = np.zeros(n) # Aquí guardaremos la sucesión de estados.
    x[0] = x0 # El primer estado es la condición inicial.
    for i in range(1,n): # Calculamos los siguientes n estados.
        x[i] = f(x[i-1],r)
    
    # Ahora calculamos el exponente de Lyapunov.
    
    sum = 0 # El profe nos dijo que la suma no incluye al término que corresponde al estado inicial.
    for i in range(1,n):
        sum = sum + np.log(abs(Df(x[i],r)))
        
    return sum/n
```

b) Ahora graficamos el exponente de Lyapunov de $f_r$ contra $r$ para varios valores de $x_0$ en $(0,1)$.

```python
x0 = 0.1
n = 1000
R = np.linspace(0.0001, 1, 1000, endpoint = False)

lya_exp_f_R = np.zeros(len(R))
for i in range(len(R)):
    lya_exp_f_R[i] = lyapunov_exp(x0,n,R[i])
    
plt.plot(R, lya_exp_f_R)
plt.xlabel("r")
plt.ylabel("Exponente de Lyapunov de f_r")
```

# Ejercicio 2

Las larvas toman dos semanas en convertirse en pupas, las pupas tardan dos semanas en convertirse en adultos y los adultos tardan dos semanas en producir larvas. Sea $L_n$, $P_n$ y $A_n$ la población de larvas, pupas y adultos en la n-ésima generación.

a) Consideremos el modelo
$$ L_{n+1} = bA_n$$
$$ P_{n+1} = (1 - c_l)L_n $$
$$ A_{n+1} = (1 - c_p)P_n + (1 - c_a)A_n $$
Interpretación de
- $b$: Es la población de larvas que cada adulto produce en cada generación.
- $L_{n+1}$: La población de larvas en la generación $n+1$ es igual a $bA_n$, la población de larvas producidas por los adultos en la generación $n$, puesto que no hay ninguna otra fuente de larvas. 
- $c_l$: Es la tasa de mortalidad de las larvas. La tasa de supervivencia de las larvas es entonces $1 - c_l$.
- $P_{n+1}$: La población de pupas en la generación $n+1$ es igual a $(1 - c_l)L_n$, la población de larvas que no muere de la generación $n$, puesto que no hay ninguna otra fuente de pupas.
- $c_p$: Es la tasa de mortalidad de las pupas. La tasa de supervivencia de las pupas es entonces $1 - c_p$.
- $c_a$: Es la tasa de mortalidad de los adultos. La tasa de supervivencia de los adultos es entonces $1 - c_a$.
- $A_{n+1}$: La población de pupas en la generación $n$ que se convierten en adultos en la generación $n+1$ es $(1 - c_p)P_n$, la población de pupas que no muere. La población de adultos en la generación $n$ que sobreviven en la generación $n+1$ es $(1 - c_a)A_n$. Como no hay ninguna otra fuente de adultos, la población de adultos en la generación $n+1$ es igual a $(1 - c_p)P_n + (1 - c_a)A_n$.

b) Consideremos el modelo
$$ L_{n+1} = bA_n\exp(-(c_{ea}A_n + c_{el}L_n))$$
$$ P_{n+1} = (1 - c_l)L_n $$
$$ A_{n+1} = (1 - c_p)P_n\exp(-c_{pa}A_n) + (1 - c_a)A_n $$
Los adultos ponen huevos, y éstos huevos se convierten en larvas. Ahora agregamos la posibilidad de que larvas y adultos se alimenten de huevos, y que adultos se alimenten de pupas. Es decir, se toman en cuenta posibles comportamientos caníbales

Interpretación de
- $c_{ea}$: Si $\exp(-c_{ea})$ es la probabilidad de que un huevo no sea comido por un adulto, entonces $\exp(-c_{ea})^{A_n}$ es la probabilidad de que un huevo no sea comido por $A_n$ adultos, suponiendo que esos eventos son independientes. $c_{ea}$ es una constante determinada experimentalmente. 
- $c_{el}$: Análogamente al caso anterior, si $\exp(-c_{el})$ es la probabilidad de que un huevo no sea comido por una larva entonces $\exp(-c_{el})^{L_n}$ es la probabilidad de que un huevo no sea comido por $L_n$ larvas, suponiendo a los eventos como independientes. $c_{el}$ es una constante determinada experimentalmente.
- $L_{n+1}$: Los adultos de la generación $n$ producen $bA_n$ huevos. La población de larvas en la generación $n+1$ es la población de huevos en la generación $n$ que sobreviven a ser comidos por adutos o larvas. Como la probabilidad de que un huevo en la generación $n$ no sea comido por ningún adulto de la generación $n$ y no sea comido por ninguna larva de la generación $n$ es $\exp(-c_{ea})^{A_n}\exp(-c_{el})^{L_n} = \exp(-(c_{ea}A_n + c_{el}L_n))$ entonces $L_{n+1} = bA_n\exp(-(c_{ea}A_n + c_{el}L_n))$
- $c_{pa}$: Si $\exp(-c_{pa})$ es la probabilidad de que una pupa no sea comida por un adulto, entonces $\exp(-c_{pa})^{A_n}$ es la probabilidad de que una pupa no sea comida por $A_n$ adultos, suponiendo que los eventos son independientos. $c_{pa}$ es una constante determinada experimentalmente.
- $A_{n+1}$: Hay dos fuentes de adultos para la generación $n+1$. La primera son los adultos de la generación $n$. Como no hay comportamientos caníbales en los que larvas, pupas o adultos coman adultos, la cantidad de adultos en la generación $n+1$ que proviene de los adultos en la generación $n$ es simplemente $(1 - c_a)A_n$, como en el modelo anterior. La segunda fuente son las pupas de la generación $n$. En este caso sí hay posibilidad de comportamientos caníbales hacia las pupas. El único comportamiento de este tipo es que los adultos coman pupas. Como la probabilidad de que una pupa en la generación $n$ no sea comida por ninguno de los adultos en esa generación es $\exp(-c_{pa})^{A_n} = \exp(-c_{pa}A_n)$ y la población de pupas en la generación $n$ que no mueren a causa de la tasa de mortalidad $c_p$ es $(1 - c_p)P_n$, entonces la población de pupas de la generación $n$ que se convierten en adultos de la generación $n+1$ es $(1 - c_p)P_n\exp(-c_{pa}A_n)$. Por lo tanto, $A_{n+1} = (1 - c_p)P_n\exp(-c_{pa}A_n) + (1 - c_a)A_n$.

c) Usando los valores $c_{ea} = 0.009$, $c_{el} = 0.012$, $c_{pa} = 0.004$, $c_l = 0.267$, $c_p = 0$ y $b = 7.48$, graficamos el diagrama de órbitas de $L_n$ con respecto al parámetro $c_a$, utilizando condiciones iniciales en el intervalo $(0,150)$.

```python
# Importamos las bibliotecas
import math
import numpy as np
import matplotlib.pyplot as plt

# En este diccionario llevamos los parámetros del sistema
parametros = {'b':7.48, 'c_ea':0.009, 'c_el':0.012, 'c_pa':0.004, 'mu_l':0.267, 'mu_p':0}

# Hacemos uso de un sistema para calcular la evolución de las poblaciones
def Sistema(larvas, crisalidas, escarabajos, parametros):
    larvas = parametros['b']*escarabajos*math.exp((-1*parametros['c_ea']*escarabajos) + (-1*parametros['c_el']*larvas))
    crisalidas = larvas*(1 - parametros['mu_l'])
    escarabajos = crisalidas*(1 - parametros['mu_p'])*math.exp(-1*parametros['c_pa']*escarabajos) + escarabajos*(1 - parametros['mu_a'])
    return larvas, crisalidas, escarabajos

# La función necesita el tiempo de relajación para
# asegurar comportamiento no transitorio, el número
# de condiciones iniciales, el valor mínimo y 
# máximo del parámetro, el número de valores del
# parámetro dentro de este intervalo, y el tamaño de
# la figura en el eje x y y
def OrbitDiag(relaxt,ptsperpar,mumin,mumax,nmu,fgszx,fgszy):
    
    # Aquí guardaremos los valores de mu para cada
    # punto de la gráfica.
    muvec=np.zeros(nmu*ptsperpar)
    
    # Aquí guardaremos los valores de la solución
    # para cada valor de mu.
    pts=np.zeros(nmu*ptsperpar)
    
    # Hacemos un vector que guarde todos los valores
    # del parametro que vamos a usar.
    muvals=np.linspace(mumin,mumax,nmu)
    
    # Contador de cuántas soluciones hemos visto.
    count=0
    
    # Loop para cada valor de mu_adultos.
    for i in range(nmu):
        
        parametros['mu_a'] = muvals[i]
        
        # Loop para cada condición inicial
        for j in range(ptsperpar):
            # Condición inicial aleatoria para larvas, crisálidas y escarabajos adultos.
            l=np.random.randint(1,150)
            c=np.random.randint(1,150)
            e=np.random.randint(1,150)
            
            # Vamos cambiando el valor de l haciendo uso de la función del sistema.
            for k in range(relaxt+int(np.floor(100*np.random.uniform(0,1)))):
                l, c, e = Sistema(l, c, e, parametros)
            # Guardamos el valor de larvas actual.
            muvec[count]=muvals[i]
            # Guardamos el valor final de la solución.
            pts[count]=l
            # Aumentamos el contador.
            count+=1
    
    # Graficamos los puntos que obtuvimos.
    plt.figure(figsize=(fgszx,fgszy), dpi=300)
    plt.scatter(muvec,pts,marker='o',s=0.01)
    plt.title('Diagrama de órbitas')
    plt.xlabel('Mortalidad de los escarabajos adultos debido al insecticida')
    plt.ylabel('Población de larvas')
    plt.show()
    
    # La función sólo grafica, no regresa nada.
    return None

# Graficamos el sistema con tiempo de relajación 100, 200 condiciones iniciales,
# valor de mu_a entre 0 y 1, con 1000 valores y tamaño de gráfica 20*10
OrbitDiag(100,200,0,1,1000,20,10)
```
